```python
year = ['2007-2008', '2008-2009', '2009-2010', '2010-2011', '2011-2012', '2012-2013', '2013-2014']
high = [3.2, 3.21, 3.25, 3.23, 3.23, 3.25, 3.24]
medhigh = [3.15, 3.16, 3.16, 3.16, 3.14, 3.17, 3.17]
medlow = [3.08, 3.07, 3.09, 3.1, 3.09, 3.09, 3.15]
low = [3.04, 3.02, 3.01, 3.04, 3.05, 3.03, 3.06]

trace0 = go.Scatter(
    x = year,
    y = high,
    name = 'High Income >$98,401',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        width = 4)
)

trace1 = go.Scatter(
    x = year,
    y = medhigh,
    name = 'Med-High Income $65,601 $98,400',
    line = dict(
        color = ('rgb(22, 96, 167)'),
        width = 4)
)

trace2 = go.Scatter(
    x = year,
    y = medlow,
    name = 'Med-Low Income $32,801 $65,600',
    line = dict(
        color = ('rgb(100, 12, 70)'),
        width = 4)
)

trace3 = go.Scatter(
    x = year,
    y = low,
    name = 'Low Income <$32,800',
    line = dict(
        color = ('rgb(40, 112, 24)'),
        width = 4)
)

data = [trace0, trace1, trace2, trace3]

layout = dict(title = 'Income and Mean GPA Over the Years',
             xaxis = dict(title = 'Year'),
             yaxis = dict(title = 'Mean GPA'),
             )

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='linegpa')
```
